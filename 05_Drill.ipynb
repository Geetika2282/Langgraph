```python
import operator
from typing import TypedDict, Annotated
from langgraph.graph import StateGraph, START, END, MessagesState
```

```python
class State(TypedDict):
    n: int
    history: Annotated[list[int], operator.add]
```

```python
def decrement(state: State) -> dict:
    decremented_value = state['n'] - 2
    return {
        'n': decremented_value,
        'history': [decremented_value]
    }
```

```python
def should_continue(state: State) -> str:
    if state['n'] <= 0:
        return END
    return 'decrement'
```

```python
builder = StateGraph(State)

builder.add_node("decrement", decrement)

builder.add_edge(START, "decrement")
builder.add_conditional_edges("decrement", should_continue, {
    "decrement": "decrement",
    END: END
})
```

```python
graph = builder.compile()
```

```python
res = graph.invoke({'n': 20, 'history': []})
print(res['history'])
```
