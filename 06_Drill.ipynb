Odd or Even routing

```python
import operator
from typing import TypedDict, Annotated
from langgraph.graph import StateGraph, MessagesState, START, END
```

```python
class State(TypedDict):
    n: int
    result: str
```

A LangGraph node must return a dictionary containing state updates.

```python
def even_node(state: State) -> dict:
    return {
        'result': 'even'
        }
```

```python
def odd_node(state: State) -> dict:
        return {
        'result': 'odd'
        }
```

```python
def text(state: State):
    if state['n'] % 2 == 0:
        return even_node
    return odd_node
```

```python
def check_number(state: State) -> str:
    if state['n'] % 2 == 0:
        return 'EVEN'
    return 'ODD'
```

```python
builder = StateGraph(State)

builder.add_node("even_node", even_node)
builder.add_node("odd_node", odd_node)
```

```python
# Route directly from START
builder.add_conditional_edges(
    START,
    check_number,
    {
        "even": "even_node",
        "odd": "odd_node"
    }
)
```

```python
# Both nodes go to END
builder.add_edge("even_node", END)
builder.add_edge("odd_node", END)
```

```python
graph = builder.compile()


final = graph.invoke({
    "n": 101,
    "result": ""
})

print(final["result"])
```
